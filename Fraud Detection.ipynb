```python
from imblearn.over_sampling import SMOTE
```

```python
X_res,y_res = SMOTE().fit_resample(X,y)
```

```python
y_res.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_res,y_res, test_size = 0.20, random_state = 42)
```

```python
log = LogisticRegression()
log.fit(X_train,y_train)
```

```python
pred1 = log.predict(X_test)
```

```python
accuracy_score(y_test,pred1)
```

```python
precision_score(y_test,pred1)
```

```python
recall_score(y_test,pred1)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
pred2 = dt.predict(X_test)
```

```python
accuracy_score(y_test,pred2)
```

```python
precision_score(y_test,pred2)
```

```python
recall_score(y_test,pred2)
```

```python
f1_score(y_test,pred2)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
```

```python
pred3 = rf.predict(X_test)
```

```python
accuracy_score(y_test,pred3)
```

```python
precision_score(y_test,pred3)
```

```python
recall_score(y_test,pred3)
```

```python
f1_score(y_test,pred3)
```

```python
final_data = pd.DataFrame({'Models':['LR','DT','RF'], 'ACC':[accuracy_score(y_test,pred1)*100,
                                               accuracy_score(y_test,pred2)*100,
                                               accuracy_score(y_test,pred3)*100]})
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the bar plot using Seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x='Models', y='ACC', data=final_data, palette='viridis')

# Adding title and labels
plt.title('Model Accuracy Comparison')
plt.xlabel('Models')
plt.ylabel('Accuracy (%)')

# Show the plot
plt.show()
```
